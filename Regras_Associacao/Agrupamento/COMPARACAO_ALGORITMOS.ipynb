# KMEANS E DBSCAN

```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
X_random, y_random = datasets.make_moons(n_samples = 1500, noise = 0.75)
```

```python
import plotly.express as px
grafico = px.scatter(x= X_random[:, 0], y = X_random[:, 1])
grafico.show()
```

```python
from sklearn.cluster import KMeans
treino = KMeans(n_clusters= 3)
centers = treino.fit_predict(X_random)
```

```python
centers
```

```python
grafico = px.scatter(x = X_random[:, 0], y = X_random[:, 1], color=centers)
grafico.show() # algoritmo eh compativel para o conjunto de dados com ruido
```

```python
np.unique(centers, return_counts=True)
```

# DBSCAN

```python
X_random, y_random = datasets.make_moons(n_samples = 1500, noise = 0)
```

```python
from sklearn.cluster import DBSCAN
```

```python
train = DBSCAN(eps = 0.5, min_samples= 5)
rotulos = train.fit_predict(X_random)
```

```python
rotulos
```

```python
np.unique(rotulos, return_counts=True) # otimo
```

```python
grafico = px.scatter(x = X_random[:, 0], y = X_random[:, 1], color = rotulos)
grafico.show() # o algoritmo nao eh compativel quando os dados possuem ruido.
```

# Agglomerative clustering HIERARCHE CLUSTERING

```python
from sklearn.cluster import AgglomerativeClustering
X_random, y_random = datasets.make_moons(n_samples = 1500, noise = 0.75)
```

```python
treino = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage='ward')
rotulos = treino.fit_predict(X_random)
```

```python
rotulos
```

```python
np.unique(rotulos, return_counts=True)
```

```python
graf = px.scatter(x = X_random[:, 0], y = X_random[:, 1], color = rotulos)
graf.show() # MELHOR QUANDO OS DADOS POSSUEM RUIDO
```
