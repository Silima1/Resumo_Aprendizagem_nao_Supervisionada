# AGRUPAMENTO HIERARQUICO COM BASE NO CARTAO DE CREDITO

```python
import pandas as pd
import numpy as np
```

```python
hc_cartao = pd.read_csv('../../../Dados/Bases_de_dados/credit_card_clients.csv', header=1)
hc_cartao
```

```python
hc_cartao['BILL_AMT_TOTAL'] = hc_cartao['BILL_AMT1'] + hc_cartao['BILL_AMT2'] + hc_cartao['BILL_AMT3'] + hc_cartao['BILL_AMT4']+ hc_cartao['BILL_AMT5'] + hc_cartao['BILL_AMT6']
```

```python
hc_cartao
```

```python
X_hc_cartao = hc_cartao.iloc[:, [1, 25]]
```

```python
X_hc_cartao
```

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler()
X_hc_cartao = scaled.fit_transform(X_hc_cartao)
X_hc_cartao
```

```python
# GERAR DENDROGRAMA PARA VERIFICAR O NUMERO DE CLUSTERS
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
Dendrograma = dendrogram(linkage(X_hc_cartao,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
HC_credito = AgglomerativeClustering(n_clusters=3, linkage='ward', affinity='euclidean')
rotulos = HC_credito.fit_predict(X_hc_cartao)
```

```python
rotulos
```

```python
import plotly.express as px
grafico = px.scatter(x = X_hc_cartao[:, 0], y = X_hc_cartao[:, 1], color=rotulos, labels{'x': 'Clientes e creditos', 'y': 'Gastos'})
grafico.show()
```
